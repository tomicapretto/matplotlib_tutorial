# Part 3: Pie Charts
Similar to bar plots we've been producing in the previous section. However, we'll be plotting the proportion of contribution from each category to the whole, instead of the absolute frequencies. 

## The basics

```python
from matplotlib import pyplot as plt
```

```python
# Values we are going to represent
slices = [60, 40]

# Plot those values in a pie chart
plt.pie(slices)
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/basic_pie_01.png')
plt.show()
```

We can see that the value `60` in our list takes the 60% of our pie, and `40` takes the rest 40%. This is not very useful because we lack of labels or guiding text in the plot. Note it is not necessary that our values add up to 100. No matter what value we have, when the propotions are calculated accordingly when the plot is produced
Let's add some extra work to the plot.

```python
# We add labels as a list
labels = ['Sixty', 'Forty']

# Note we don't add `plt.legend()`. In that case an extra legend would appear.
plt.pie(slices, labels=labels)
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/basic_pie_02.png')
plt.show()
```

If we wanted to change the color of the line separating each piece of the pie, we can do that by specifying the `edgecolor` in the `wedgeprops` argument.  
Note we have to pass a dictionary.  

For more documentation on wedge properties [click here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.patches.Wedge.html).

```python
plt.pie(slices, labels=labels, wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/basic_pie_03.png')
plt.show()
```

In addition, we can change the colors of the slices. The logic is similar to what we did for the the labels. We just create a list with the colors we want to pass to each category. Let's add some extra slices so it is more interesting.

```python
slices = [120, 80, 30, 20]
labels = ['Sixty', 'Forty', 'Extra1', 'Extra2']
colors = ['blue', 'red', 'yellow', 'green']

plt.pie(slices, labels=labels, colors=colors,
        wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/basic_pie_04.png')
plt.show()
```

Calling the colors by their names is something easy and straightforward. However the colors obtained are not very atractive (too bright).  
More and better possibilities can be incorporated by representing colors via HEX values. Custom values can be found in online palettes like the one [here](https://htmlcolorcodes.com).  
Let's see how it works.

```python
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']

plt.pie(slices, labels=labels, colors=colors,
        wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/basic_pie_05.png')
plt.show()
```

## Pie charts to plot the most popular programming langauges
This works with the same dataset used in part 02. No data importing and handling is performed here because it was already done. We only copy two lists that represent the counts and the programming languages.

```python
# Language Popularity. Data from stackoverflow developer survey. Top 15 langauges.
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 
          'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels=labels,
        wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/proglang_pie_01.png')
plt.show()
```

In my opinion this pie chart is an excellent example. An excellent example of how you must not use a pie chart. With so many categories (above 5/6) pie charts look terribly bad and are not that easy/straightforward to read.   

I think a good pie chart is one that shows summary information very easily and quickly. A pie chart like the one shown here does not meeet that criteria.

If you still want to plot as many categories, you should use the bar chart we showed in previous section.

```python
# Now let's keep top 5 languages.

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices, labels=labels,
        wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/proglang_pie_02.png')
plt.show()
```

Now let's see what else we can do with pie charts. For example we can emphasize on a specific slice of the pie. Suppose we go to a Python conference and we want to emphasize this category.  

We'll use a list called `explode`. This contains floats that specifies how much we want to emphasize that slice. 
For example, 0 leaves the slice as it is. Any value above 0 specifies the fraction of the radious we want to explode the value. If it does not make sense, let's take a look.

```python
explode = [0, 0, 0, 0.1, 0]  # python is in the 4th position.

plt.pie(slices, labels=labels, explode=explode,
        wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/proglang_pie_03.png')
plt.show()
```

Now, Python is exploded out from the pie. This adds emphasis to Python category. The larger the explode value, the more distance between the explode slice and the rest of the pie.  

Explode values are in terms of proportion of the radious of the pie. As we see in the next plot, the point where Python slice is located is at half of the radious (explode = 0.5)

```python
# For example, this would look very weird.
explode = [0, 0, 0, 0.5, 0]

plt.pie(slices, labels=labels, explode=explode,
        wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/proglang_pie_04.png')
plt.show()
```

There are a few more arguments that we can pass to the `pie` method that can be useful. Let's have a look at some of these.

```python
explode = [0, 0, 0, 0.1, 0] # go back to acceptable explode

# Adds shadows and look more like 3D stuff.

plt.pie(slices, labels=labels, explode=explode, shadow=True,
        wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/proglang_pie_05.png')
plt.show()
```

```python
# change start angle, which is like rotating the original pie.

plt.pie(slices, labels=labels, explode=explode, shadow=True, startangle=90, # 0 by default.
        wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/proglang_pie_06.png')
plt.show()
```

```python
# Display percentage of each slice

plt.pie(slices, labels=labels, explode=explode, shadow=True, 
        startangle=90, autopct='%1.1f%%',  # format string
        wedgeprops={'edgecolor':'black'})
plt.title('My awesome Pie Chart')
plt.tight_layout()
plt.savefig('output/imgs/part_03/proglang_pie_07.png')
plt.show()
```
