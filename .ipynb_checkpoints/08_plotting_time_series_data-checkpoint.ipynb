# Part 8: Plotting Time Series Data
First, we are going to look at basic examples based on some toy data. Then, we are going to work with bitcoin prices imported from a csv file.

```python
import pandas as pd
from datetime import datetime, timedelta   # new module
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates  # new module
```

```python
# List of 7 dates. Requires datetime module.
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

# Values
y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates, y)
plt.tight_layout()
plt.show()
```

We can see there's a warning (there's no need to worry much about that now) and that the default plot shows markers instead of lines. This can be changed.

```python
plt.plot_date(dates, y, linestyle='solid')
plt.tight_layout()
plt.show()
```

Now the markers are connected by a line, but this does not delete the markers. If we wanted to delete them we can set `marker='none'`.  
It is possible to automatically format the dates on the x-axis to make them fit better. The method that is going to be called is called on a figure, not the pyplot object.

To get the current figure from pyplot we use `plt.gcf()`.

```python
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()
plt.tight_layout()
plt.show()
```

What if we wanted to change the format of the dates? What if we wanted month-day-year instead of year-month-day?  
This requires using some datetime formatting. This is why we imprted `dates` from `matplotlib` in the very beginning.

We are going to use the `DateFormatter` class. More information on how to work with datetimes can be found [here](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior).

```python
# Desired format
date_format = mpl_dates.DateFormatter('%b, %d %Y') # The "," is intentional.
```

We need to set this format as our format to our x-axis. As we grab the figure to run the autofmt method, we are going to grab the axis to run this other format method.

```python
plt.plot_date(dates, y, linestyle='solid')
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(date_format)
plt.tight_layout()
plt.show()
```

Now we are ready to move to some real data to see how to work with data imported from a csv file.

```python
data = pd.read_csv('data/part_08/data.csv')
price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Price')
plt.xlabel('Date')
plt.ylabel('Close price')

plt.tight_layout()
plt.show()
```

If it is the case that our Date column is being imported as string instead of datetime, we can fix that by using `to_datetime()` method. In addition, it is always good to sort the datetimes to ensure they are plotted in the proper order.

```python
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)

price_date = data['Date']

plt.plot_date(price_date, price_close, linestyle='solid')
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Price')
plt.xlabel('Date')
plt.ylabel('Close price')

plt.tight_layout()
plt.show()
```
