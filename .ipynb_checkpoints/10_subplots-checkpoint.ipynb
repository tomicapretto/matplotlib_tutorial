# Part 10: Subplots
We are going to see how to create plots using the `subplots` method.

```python
import pandas as pd
from matplotlib import pyplot as plt
```

The csv file contains the same data we used for the first of the tutorials. It is about developers median salaries by age.

The following code and resulting plot is familiar from the previous parts.

```python
data = pd.read_csv('data/part_10/data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, py_salaries, label="Python")
plt.plot(ages, js_salaries, label="JavaScript")
plt.plot(ages, dev_salaries, color="#444444", 
         linestyle="--", label="All Devs")
plt.legend()

plt.title("Median Salary by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

Every time we are doing something with a plot we are using the `plt` object that we imported. We are never creating that object. This object has a current state in terms of which figure we are working with and what axis we are working with (and so on). We've been working with single figure and single axis objects.

We can think of the figure as the container holding our plots. We can think as the whole window that appears when we run our code. The axes are the actual plots. Thus, the figure can contain multiple plots.

Many people (Corey included) prefer to use the `subplots` method to work in a more object-oriented style.

Right now, let's simply recreated the plot we've just created. We are going to do this by creating our own figure and axis.

```python
fig, ax = plt.subplots() # If we don't pass anything, it is just a 1 row and 1 col plot.

# Here we just replace the `plt.plot` with `ax.plot`
ax.plot(ages, py_salaries, label="Python")
ax.plot(ages, js_salaries, label="JavaScript")
ax.plot(ages, dev_salaries, color="#444444", 
         linestyle="--", label="All Devs")

ax.legend()

# When it comes to titles and labes, this is a little different.
# We change `plt` to `ax` but we use the "set_*". 
ax.set_title("Median Salary by Age")
ax.set_xlabel("Ages")
ax.set_ylabel("Median Salary (USD)")

# These two don't change
plt.tight_layout()
plt.show()
```

Let's say we want the Python and JavaScript salaries together in a plot and All Devs in a separated plot. We can just do that by adding an extra axis.

The first argument to the `subtplots` method refer to the rows and columns of the layout created.

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
print(ax) 
```

We see the `axis` object is a list that contains two AxesSubplot objects. These two axis will live in the same figure. 

Another way is unpacking the list of axis.

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
print(ax1)
print(ax2)
```

We just got two AxesSubplot individually.

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(ages, dev_salaries, color="#444444", 
          linestyle="--", label="All Devs")

ax2.plot(ages, py_salaries, label="Python")
ax2.plot(ages, js_salaries, label="JavaScript")


ax1.legend() 
ax1.set_title("Median Salary by Age")
ax1.set_xlabel("Ages")
ax1.set_ylabel("Median Salary (USD)")

ax2.legend() 
ax2.set_title("Median Salary by Age")
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

A couple of things:
- We don't need a title for each single plot. We could only use the title on top.
- Labelling the x axis in both plots is also unnecesary. We could only label the bottom x axis.
- If we wanted, we could only mark the tick axes in one of the plots. This is achieved by the `sharex` argument in `subplots` method.

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(ages, dev_salaries, color="#444444", 
          linestyle="--", label="All Devs")

ax2.plot(ages, py_salaries, label="Python")
ax2.plot(ages, js_salaries, label="JavaScript")


ax1.legend() 
ax1.set_title("Median Salary by Age")
ax1.set_xlabel("Ages")
ax1.set_ylabel("Median Salary (USD)")

ax2.legend() 
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

What if we wanted to have these plots in two different figures?  
Very simple:

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color="#444444", 
          linestyle="--", label="All Devs")

ax2.plot(ages, py_salaries, label="Python")
ax2.plot(ages, js_salaries, label="JavaScript")

ax1.legend() 
ax1.set_title("Median Salary by Age")
ax1.set_xlabel("Ages")
ax1.set_ylabel("Median Salary (USD)")

ax2.legend() 
ax2.set_title("Median Salary by Age")
ax2.set_xlabel("Ages")
ax2.set_ylabel("Median Salary (USD)")

plt.tight_layout()
plt.show()
```

If we wanted to save each of thes figures, we can still use the `savefig` method with each of the figures.

```python
fig1.savefig('output/imgs/part_10/fig1.png')
fig2.savefig('output/imgs/part_10/fig2.png')
```
