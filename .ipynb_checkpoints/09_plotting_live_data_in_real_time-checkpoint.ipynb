# Part 9: Plotting live data in real time
This type of plots arises when we have data changing frequently and we want to monitor.

```python
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

First we are going to use some toy data that is on Python to see how this works. After it, we are going to plot data from a csv file. We will be monitoring the csv file for changes, and when something changes, the plot will be updated.

```python
x_vals = [0, 1, 2, 3, 4, 5]
y_vals = [0, 1, 3, 2, 3, 5]
```

```python
# The following is a static line chart.

plt.plot(x_vals, y_vals)
plt.tight_layout()
plt.show()
```

Now let's delete the values from the lists `x_vals` and `y_vals` and create a function that appends new data to each list. 

The `count()` function just counts one value at a time, and each time you call `next()` you get the next value.

```python
x_vals = []
y_vals = []

index = count()

# This function appends values to both x_vals and y_vals
# The values appended to x_vals are just indexes (counts up by 1)
# while values appended to y_vals are random integer between 0 and 5.

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    
    plt.plot(x_vals, y_vals)
```

Suppose we want to run this function every second a plot these values being appended to our list. To do that we can use the `FuncAnimation` class from the matplotlib animation module.

```python
# We have to pass the figure we want to animate, the function we want to run for the animation
# and the interval for how often we want to run it (in ms).

# NOTE: It does not work in Jupyter Notebook without the notebook backend.
%matplotlib notebook

ani = FuncAnimation(plt.gcf(), animate, interval=1000) # gcf = get current figure

plt.tight_layout()
plt.show()
```

The problem with the current implementation is that a new line is being plotted each time the function `animate()` is being called (without cleaning the old line). That explains why the color of the line changes all the time. We don't note the other colors because each new line is being plotted above the others.

```python
x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    
    plt.cla()  # clear axis function
    plt.plot(x_vals, y_vals)

ani = FuncAnimation(plt.gcf(), animate, interval=1000) # gcf = get current figure

plt.tight_layout()
plt.show()
```

Now let's see how this works in real life. We are going to plot data that is being stored in a csv file. This file is constantly being updated from an outside source. We have such files when we are constantly exporting some data from an online source, data from API, data from sensors, etc...

In our case we have a Python script that is generating random data all the time and it is being saved in a csv file.  

The script is stored in the same repository under the name of `data_generator.py` and is run from the terminal.

```python
def animate(i):
    data = pd.read_csv('data/part_09/data.csv')
    x = data['x_value']
    y1 = data['total_1']
    y2 = data['total_2']
    
    plt.cla()
    plt.plot(x, y1, label="Channel 1")
    plt.plot(x, y2, label="Channel 2")
    
    plt.legend(loc='upper left') # plt.cla() also clears legends
    plt.tight_layout()

ani = FuncAnimation(plt.gcf(), animate, interval=1000)
plt.tight_layout()
plt.show()
```
