# Part 7: Scatterplots
Scatterplots are useful when you have two variables and want to see their relationship in a plot. They are useful to detect where two variables are correlated.

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')
```

## 1. Basic example
We are going to create a very basic scatterlpot using some fake data in python lists.

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y)

plt.tight_layout()
plt.show()
```

We know we have random values in both `x` and `y`, and this is confirmed in the scatterplot, where no trend or pattern is seen.  

Let's look at some modifications we can do to the scatterplot.

```python
plt.scatter(x, y, s=100, c='green', marker='X') # `s`= `size`, `c` = `color`.

plt.tight_layout()
plt.show()
```

Matplotlib contains a lot of different options for `markers`. Official documentation on how to customize these formats can be found [here](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html).

If we want to improve the appearence of the scatterplots, we can add edges to the circles and some transparency to the fill of the markers by passing an `alpha` argument.

```python
plt.scatter(x, y, s=100, c='green', 
            edgecolor='black', linewidth=1, alpha=0.6)

plt.tight_layout()
plt.show()
```

It is also possible to add specific colors to the different markers, depending on the value of another variable. Size can be modified in this manner, too.

```python
# this list can be a satisfaction measurement
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5] #between 1 and 10.

plt.scatter(x, y, s=100, c=colors, 
            edgecolor='black', linewidth=1, alpha=0.6)

plt.tight_layout()
plt.show()
```

We see we get different color dots, each point receives a different shade of grey, depending on its `color` value.  
If we want to change this grey colormap, we can use other custom colormap available in matplotlib. As with the markers, there are a bunch of colormaps available. They can be found [here](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html).

```python
plt.scatter(x, y, s=100, c=colors, cmap='Greens',       # `cmap` = `Color Maps`.
            edgecolor='black', linewidth=1, alpha=0.6)

plt.tight_layout()
plt.show()
```

The plot looks much better, but we still don't know what is being indicated by the intensity of the color. We can add a label to the colormap.

```python
plt.scatter(x, y, s=100, c=colors, cmap='Greens',
            edgecolor='black', linewidth=1, alpha=0.6)

# Add a colorbar to the plot.
cbar = plt.colorbar()
cbar.set_label('Satisfaction level')

plt.tight_layout()
plt.show()
```

Finally, let's see how we can add size to the dots. Usually they represent population size or sample size being represented by each point.

```python
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538,
        239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s=sizes, c=colors, cmap='Greens',
            edgecolor='black', linewidth=1, alpha=0.6)

# Add a colorbar to the plot.
cbar = plt.colorbar()
cbar.set_label('Satisfaction level')

plt.tight_layout()
plt.show()
```

## 2. Real data.
Data is related to top 200 youtube trend videos on the day Corey recorded the video related to this tutorial.  
We'll be plotting a scatterplot of the total views and the total likes. We also have calculated the ratio of likes/dislikes as well.

```python
data = pd.read_csv('data/part_07/data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

We see that the vast majority of dots are placed on the bottom-left part of the plot, while there is an outlier video that has much larger View Count and Total Likes than the rest of the videos.  

Overall, there is a clear positive relationship between View Count and Total Likes.

Let's see how the plot changes when we use a logarithm scale for both variables.

```python
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

# Change x and y scales.
plt.xscale('log') 
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

The positive correlation that we detected in the previous scatterplot is much clearer in the log-scale.  
Let's add the ratio of likes/dislike to the plot. We will include it as colors. It wouldn't be a good idea to add it as sizes because the dots are very close to each other.

```python
plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

# Change x and y scales.
plt.xscale('log') 
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()
plt.show()
```

In the plot we can see that those with a color more green/blue performed less well in the like/dislike ratio sense, while the ones in yellow performed better.
